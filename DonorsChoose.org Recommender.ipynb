```python
def splitDataFrameList(df,target_column,separator):
    ''' df = dataframe to split,
    target_column = the column containing the values to split
    separator = the symbol used to perform the split
    returns: a dataframe with each entry for the target column separated, with each element moved into a new row. 
    The values in the other columns are duplicated across the newly divided rows.
    '''
    row_accumulator = []

    def splitListToRows(row, separator):
        split_row = row[target_column].split(separator)
        for s in split_row:
            new_row = row.to_dict()
            new_row[target_column] = s
            row_accumulator.append(new_row)

    df.apply(splitListToRows, axis=1, args = (separator, ))
    new_df = pd.DataFrame(row_accumulator)
    return new_df
```

Importing tfidf vectorizer and linear kernel libraries for further evaluation. Tfidf vectorizer will generate a sparse matrix that contains the term frequency in a sentence and inverses the frequency of that term in all the sentences. Linear Kernel will be helpful in calculating the cosine similarity of this calculated tfidf matrix.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
from sklearn.metrics.pairwise import linear_kernel
```

```python
def comp(let):
    
    #code for donor based part of the recommender sytem
    #the code is self explanatory
    city=Donors[Donors['Donor ID']==let]['Donor City'].values[0]
    state=Donors[Donors['Donor ID']==let]['Donor State'].values[0]
    teacher=Donors[Donors['Donor ID']==let]['Donor Is Teacher'].values[0]
    city=Donors[Donors['Donor ID']==let]['Donor City'].values[0]
    zipp=Donors[Donors['Donor ID']==let]['Donor Zip'].values[0]
    c=Donors[Donors['Donor City']==city]
    s=Donors[Donors['Donor State']==state]
    t=Donors[Donors['Donor Is Teacher']==teacher]
    z=Donors[Donors['Donor Zip']==zipp]
    com=pd.concat([c,s,t,z])
    comb=com.drop_duplicates(subset=['Donor City','Donor State','Donor Is Teacher','Donor Zip'])
    cs=[]
    for csos in comb['Donor City'].values:
        if csos==city:
            cs.append(4)
        else:
            cs.append(0)
    comb['cs']=cs
    ss=[]
    for csos in comb['Donor State'].values:
        if csos==state:
            ss.append(2)
        else:
            ss.append(0)
    comb['ss']=ss
    ts=[]
    for csos in comb['Donor Is Teacher'].values:
        if csos==teacher:
            ts.append(2)
        else:
            ts.append(0)
    comb['ts']=ts
    zs=[]
    for csos in comb['Donor Zip'].values:
        if csos==zipp:
            zs.append(2)
        else:
            zs.append(0)
    comb['zs']=zs
    comb.reset_index(inplace=True)
    comb.drop('index',axis=1,inplace=True)
    comb['tot']=comb['cs']+comb['ss']+comb['ts']+comb['zs']
    comb.sort_values(by='tot',ascending=False,inplace=True)
    comb.reset_index(inplace=True)
    comb.drop('index',axis=1,inplace=True)
    #capping the score to 6, the donors having scores below 6 will be deleted
    comb=comb[(comb['tot']>=6)] 
    #code for getting the projects of the donors for further application of project based part of the recommendation sytem
    lstd=comb['Donor ID'].values
    lstp=Donations[Donations['Donor ID'].isin(lstd)==True]['Project ID'].values
    comb=proj2[proj2['Project ID'].isin(lstp)==True]
    comb.reset_index(inplace=True)
    comb.drop('index',axis=1,inplace=True)
    
    #project based part of the recommendation system for different columns
    #project short description
    tfidf_descr = tfidf.fit_transform(comb['Project Short Description'])
    tf_idf_descr=tfidf_descr.todense()
    tf_idf_descr=pd.DataFrame(tf_idf_descr)
    cosine_sim_descr = linear_kernel(tf_idf_descr, tf_idf_descr)
    indices_descr = pd.Series(comb.index, index=comb['Project ID'])
    tel=Donations[Donations['Donor ID']==let]['Project ID'].values[0]
    idx_descr = indices_descr[indices_descr.index==tel].values[0]
    sim_scores_descr = list(enumerate(cosine_sim_descr[idx_descr]))
    sim_scores_descr = sorted(sim_scores_descr, key=lambda x: x[1], reverse=True)
    proj_indices_descr = [i[0] for i in sim_scores_descr]
    descr=comb.iloc[proj_indices_descr][1:]
    descr.reset_index(inplace=True)
    descr.drop('index',axis=1,inplace=True)
    descr.reset_index(inplace=True)
    descr['index']=len(descr)-1-descr['index']
    descr['score1']=descr['index']
    descr.drop('index',axis=1,inplace=True)
    
    #project need statement
    tfidf_need = tfidf.fit_transform(comb['Project Need Statement'])
    tf_idf_need=tfidf_need.todense()
    tf_idf_need=pd.DataFrame(tf_idf_need)
    cosine_sim_need = linear_kernel(tf_idf_need, tf_idf_need)
    indices_need = pd.Series(comb.index, index=comb['Project ID'])
    tel=Donations[Donations['Donor ID']==let]['Project ID'].values[0]
    idx_need = indices_need[indices_need.index==tel].values[0]
    sim_scores_need = list(enumerate(cosine_sim_need[idx_need]))
    sim_scores_need = sorted(sim_scores_need, key=lambda x: x[1], reverse=True)
    proj_indices_need = [i[0] for i in sim_scores_need]
    need=comb.iloc[proj_indices_need][1:]
    need.reset_index(inplace=True)
    need.drop('index',axis=1,inplace=True)
    need.reset_index(inplace=True)
    need['index']=len(need)-1-need['index']
    need['score2']=need['index']
    need.drop('index',axis=1,inplace=True)
    
    #soup
    tfidf_soup = tfidf.fit_transform(comb['soup'])
    tf_idf_soup=tfidf_soup.todense()
    tf_idf_soup=pd.DataFrame(tf_idf_soup)
    cosine_sim_soup = linear_kernel(tf_idf_soup, tf_idf_soup)
    indices_soup = pd.Series(comb.index, index=comb['Project ID'])
    tel=Donations[Donations['Donor ID']==let]['Project ID'].values[0]
    idx_soup = indices_soup[indices_soup.index==tel].values[0]
    sim_scores_soup = list(enumerate(cosine_sim_soup[idx_soup]))
    sim_scores_soup = sorted(sim_scores_soup, key=lambda x: x[1], reverse=True)
    proj_indices_soup = [i[0] for i in sim_scores_soup]
    soup=comb.iloc[proj_indices_soup][1:]
    soup.reset_index(inplace=True)
    soup.drop('index',axis=1,inplace=True)
    soup.reset_index(inplace=True)
    soup['index']=len(soup)-1-soup['index']
    soup['score3']=soup['index']
    soup.drop('index',axis=1,inplace=True)
    
    #code for joining and aggregating these dataframes produced by different columns    
    df=pd.merge(descr,need,how='outer',on=['Project ID','Project Title','Project Short Description','Project Need Statement','soup'])
    fin1=pd.merge(df,soup,how='outer',on=['Project ID','Project Title','Project Short Description','Project Need Statement','soup'])
    fin1.fillna(0,inplace=True)
    fin1['tot']=fin1['score1']+fin1['score2']+fin1['score3']
    fin1.sort_values(by='tot',inplace=True,ascending=False)
    x=Donations[Donations['Donor ID']==let]['Project ID'].values[0]
    y=fin1[fin1['Project ID']==x]['Project ID'].values[0]
    fin2=fin1[fin1['Project ID']!=y]
    fin2.reset_index(inplace=True)
    fin2.drop('index',axis=1,inplace=True)
    lst=list(fin2['Project ID'].values)
    fin3=Projects[Projects['Project ID'].isin(lst)==True]
    fin2.sort_values(by='Project ID',inplace=True)
    fin2.reset_index(inplace=True)
    fin2.drop('index',axis=1,inplace=True)
    fin3.sort_values(by='Project ID',inplace=True)
    fin3.reset_index(inplace=True)
    fin3.drop('index',axis=1,inplace=True)
    fin3['tot']=fin2['tot']
    fin3.sort_values(by='tot',inplace=True,ascending=False)
    fin3.reset_index(inplace=True)
    fin3.drop('index',axis=1,inplace=True)
    
    return fin3
```

Method for getting the score of the Project IDs, can be used by other columns too. Score is just their index reversed in order.

```python
def rank_call(df):
    pf=df.copy()
    temp1=pd.DataFrame(pf['Project ID'])
    temp1.reset_index(inplace=True)
    temp1.drop('index',axis=1,inplace=True)
    temp1.reset_index(inplace=True)
    temp1['index']=len(temp1)-temp1['index']
    temp1['rank1']=temp1['index']
    temp1.drop('index',axis=1,inplace=True)
    return temp1
```

This is the method that aggregates the result of both the simple recommender and project and donor based recommender. We need to call just this method passing in the the donor id or the information of the donor. This will give the final result of this complete recommender system.
