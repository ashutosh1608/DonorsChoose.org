```python
def categorize_age(age):
    if age == '0-17':
        return 1
    elif age == '18-25':
        return 2
    elif age == '26-35':
        return 3
    elif age == '36-45':
        return 4
    elif age == '46-50':
        return 5
    elif age == '51-55':
        return 6
    elif age == '55+':
        return 7
```

```python
trainer['age_categ']=trainer['Age'].apply(lambda x: categorize_age(x))
```

```python
trainer.drop('Age',axis=1,inplace=True)
trainer.head()
```
