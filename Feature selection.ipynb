```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
trainer=pd.read_csv('train.csv')
```

```python
trainer.head()
```

```python
trainer.drop(['User_ID','Product_ID'],axis=1,inplace=True)
```

```python
trainer.head()
```

```python
def categorize_age(age):
    if age == '0-17':
        return 1
    elif age == '18-25':
        return 2
    elif age == '26-35':
        return 3
    elif age == '36-45':
        return 4
    elif age == '46-50':
        return 5
    elif age == '51-55':
        return 6
    elif age == '55+':
        return 7
```

```python
trainer['age_category']=trainer['Age'].apply(lambda x: categorize_age(x))
```

```python
trainer.head()
```

```python
trainer.drop('Age',axis=1,inplace=True)
```

```python
trainer.head()
```

```python
gender=pd.get_dummies(trainer['Gender'],drop_first=True)
```

```python
trainer.drop('Gender',axis=1,inplace=True)
```

```python
trainer=pd.concat([trainer,gender],axis=1)
```

```python
trainer.head()
```

```python
def categorize_city(city):
    if city=='A':
        return 1
    elif city=='B':
        return 2
    elif city=='C':
        return 3
```

```python
city=trainer['City_Category'].apply(lambda x: categorize_city(x))
```

```python
city
```

```python
trainer['city']=city
```

```python
trainer.head()
```

```python
trainer.drop('City_Category',axis=1,inplace=True)
```

```python
trainer.head()
```

```python
trainer.fillna(0.0,inplace=True)
```

```python
trainer.head()
```

```python
def city_years(years):
    if years=='4+':
        return 4
    return years
```

```python
trainer['city_years']=trainer['Stay_In_Current_City_Years'].apply(lambda x: city_years(x))
```

```python
trainer.drop('Stay_In_Current_City_Years',axis=1,inplace=True)
```
